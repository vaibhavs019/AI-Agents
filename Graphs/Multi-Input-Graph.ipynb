```python
from typing import TypedDict, List
from langgraph.graph import StateGraph
from numpy.ma.core import product
```

```python
class AgentState(TypedDict):
    values: List[int]
    name: str
    result: str
```

```python
def process_values(state: AgentState) -> AgentState:
    """This functions handles multiple different inputs"""
    print(f"Initial state: {state}")
    state['result'] = f"Hi, {state['name']}! The sum of the values is {sum(state['values'])}"
    print(f"Final state: {state}")
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("process", process_values)
# Setting entry and finish points
graph.set_entry_point("process")
graph.set_finish_point("process")
app = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({'values': [1, 2, 3], 'name': 'Vaibhav'})
```

```python
from typing import TypedDict, List
from langgraph.graph import StateGraph
from numpy.ma.core import product
```

```python
class AgentState(TypedDict):
    values: List[int]
    name: str
    result: str
    operation: str
```

```python
def process_values(state: AgentState) -> AgentState:
    """This functions calculates sum or product"""
    print(f"Initial state: {state}")
    if state['operation'] == '+':
        state['result'] = f"Hi, {state['name']}! The sum of the values is {sum(state['values'])}"

    elif state['operation'] == '*':
        state['result'] = f"Hi, {state['name']}! The sum of the values is {product(state['values'])}"

    print(f"Final state: {state}")
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("process", process_values)
# Setting entry and finish points
graph.set_entry_point("process")
graph.set_finish_point("process")
app = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({'values': [1, 2, 3], 'name': 'Vaibhav', 'operation': '*'})
```
